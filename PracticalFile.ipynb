```python
import math

print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 4 :")
print("Program to print sum of all the primes below two million : ")

def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

total = 0
for num in range(2, 2000000):
    if is_prime(num):
        total += num

print("Sum of all primes below 2 million is : ", total)

```

```python
print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 7 :")
print("Program to count characters in a string and store in a dictionary: ")

text = input("Enter a string : ")

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print("Character count dictionary : ")
print(char_count)
```

```python
print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 10 :")
print("Program to show the multiple inheritance concepts : ")

class Parent1:
    def parent1_info(self):
        print("This is Parent 1 class.")

class Parent2:
    def parent2_info(self):
        print("This is Parent 2 class.")

class Child(Parent1, Parent2):
    def child_info(self):
        print("This is the Child class.")

c = Child()
c.parent1_info()
c.parent2_info()
c.child_info()
```

```python
print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 11 :")
print("Program to perform insert, update, delete and retrieve the data : ")

import sqlite3

# Create in-memory database
con = sqlite3.connect(':memory:')
cur = con.cursor()

# Create table
cur.execute("CREATE TABLE student(id INTEGER, name TEXT)")

# Insert data
cur.execute("INSERT INTO student VALUES(1, 'Rahul')")

# Update data
cur.execute("UPDATE student SET name='Amit' WHERE id=1")

# Retrieve data
cur.execute("SELECT * FROM student")
print("Data after update:", cur.fetchall())

# Delete data
cur.execute("DELETE FROM student WHERE id=1")

# Commit changes and close connection
con.commit()
con.close()
```

```python
print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 15 :")
print("Program to clean a CSV file for missing values : ")

import pandas as pd

try:
    data = pd.read_csv("data.csv")
    print("Original Data : \n", data)

    # Fill missing values with 'Unknown'
    data_filled = data.fillna("Unknown")
    print("\nCleaned Data :\n", data_filled)

    # Save the cleaned data to a new CSV file
    data_filled.to_csv("cleaned_data.csv", index=False)
    print("\nCleaning complete! 'cleaned_data.csv' created successfully.")
except FileNotFoundError:
    print("Error: 'data.csv' file not found.")
except Exception as e:
    print("An error occurred:", e)
```

```python
print("Name : Manoj Singh")
print("Course : MCA-1st Sem")
print("Program 17 :")
print("Program for stock market prediction : ")

import numpy as np
from sklearn.linear_model import LinearRegression

# Training data: days vs stock prices
days = np.array([[1], [2], [3], [4], [5]])
price = np.array([100, 105, 109, 110, 111])

# Create and train the model
model = LinearRegression()
model.fit(days, price)

# Predict stock price for day 6
pred = model.predict([[6]])
print("Predicted stock price on day 6 : ", pred[0])
```
