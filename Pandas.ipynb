```python
# Aggregation
import pandas as pd

# Sample Data
data = {
    "Department": ['HR', 'Finance', 'IT', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    "Employee": ['Ram', 'Shyam', 'Ghanshyam', 'Aditi', 'Sachin', 'Mayank', 'Raj', 'Rahul'],
    "Age": [25, 30, 28, 35, 29, 32, 40, 31],
    "Salary": [50000, 60000, 55000, 70000, 65000, 52000, 72000, 68000],
    "Performance_Score": [80, 90, 85, 95, 88, 82, 96, 89]
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)
print(df['Salary'].mean())
print(df['Salary'].sum())
print(df['Performance_Score'].min())
print(df['Age'].max())
```

```python
# Grouping
import pandas as pd

# Sample Data
data = {
    "Department": ['HR', 'Finance', 'IT', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    "Employee": ['Ram', 'Shyam', 'Ghanshyam', 'Aditi', 'Sachin', 'Mayank', 'Raj', 'Rahul'],
    "Age": [25, 25, 28, 35, 29, 31, 40, 31],
    "Salary": [50000, 60000, 55000, 70000, 65000, 52000, 72000, 68000],
    "Performance_Score": [80, 90, 85, 95, 88, 82, 96, 89]
}
df=pd.DataFrame(data)
print(df)
grp=df.groupby("Age")['Salary'].sum()
print(grp)
#Multiple Aggregations
grouped=df.groupby(['Age','Employee'])["Salary"].sum()
print(grouped)
```

```python
# Merging and Joining
import pandas as pd
df_customers=pd.DataFrame({
    'CustomerID':[1,2,3],
    'Name':['Ram','Shyam','Ghanshyam']
})
df_order=pd.DataFrame({
    'CustomerID':[1,2,4],
    'OrderAmount':[250,450,350]
})
df_merged=pd.merge(df_customers,df_order,on='CustomerID',how='inner')
print("Inner Joining")
print(df_merged)
df_merged=pd.merge(df_customers,df_order,on='CustomerID',how='outer')
print("Outer Joining")
print(df_merged)
df_merged=pd.merge(df_customers,df_order,on='CustomerID',how='left')
print("Left Joining")
print(df_merged)
df_merged=pd.merge(df_customers,df_order,on='CustomerID',how='right')
print("Right Joining")
print(df_merged)
```

```python
#Concatenation
import pandas as pd
df1 = pd.DataFrame({
    'CustomerID': [1, 2],
    'Name': ['Ram', 'Shyam']
})

df2 = pd.DataFrame({
    'CustomerID': [3, 4],
    'Name': ['Ghanshyam', 'Aditi']
})

result = pd.concat([df1, df2], ignore_index=True)
print(result)
result = pd.concat([df1, df2],axis=1, ignore_index=True)
print('Horizontal')
print(result)
```
