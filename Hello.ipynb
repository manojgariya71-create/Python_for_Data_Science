```python
#if else Condition
age=int(input("Enter your age"))
if(age>=18):
    print("You are eligible for vote")
else:
    print("you are not eligible for vote")
```

```python
# elif else Condition
age=int(input("Enter your age"))
if(age>=18 and age<=100):
    print("You are eligible for vote")
elif(age<=0):
    print("Enter the valible age")
elif(age>=101):
    print("You are greater than 101")
else:
    print("You are not eligible for vote")
```
