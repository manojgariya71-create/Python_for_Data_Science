‚úÖ What is a Function?

A function is a block of reusable code that performs a specific task.
Instead of writing the same code again and again, we write it once in a function and call it whenever needed.

‚úÖ Why do we use Functions?

To avoid repeating code

To make programs modular and organized

To make code easier to understand and debug

To reuse logic multiple times

‚úÖ Types of Arguments in Functions

When you call a function, the values you pass inside it are called arguments.

1Ô∏è‚É£ Positional Arguments

‚úî Order matters
‚úî Values are matched by position

2Ô∏è‚É£ Keyword Arguments

‚úî Order does NOT matter
‚úî You use the parameter name while passing values

3Ô∏è‚É£ Default Arguments

‚úî Function has a default value
‚úî If no value is passed, the default is used

```python
# positional arguments
def showdata(name,city):
    print("Hello", name +" "+"welcome to the "+city)
showdata("Rohit","Dehradun")
```

```python
# keyword arguments
def showdata(name,city):
    print("Hello", name +" "+"welcome to the "+city)
showdata(name="Manoj",city="Mumbai")
```

```python
# default arguments
def showdata(name="Priya",city="Delhi"):
    print("Hello", name +" "+"welcome to the "+city)
showdata()
showdata(name="Shubham",city="Dehradun")
```

‚úÖ What is a return statement?

A return statement is used inside a function to send a value back to the place where the function was called.

In simple words:

üëâ return = function ka answer
üëâ Function stops running as soon as it hits return

‚úÖ Why do we use return?

To give output from a function

To store the result in a variable

To use the result later in the program

```python
def full_name(first_name,last_name):
    return first_name+" "+last_name
name=full_name("Manoj","Singh")
print(name)
```

‚úÖ Local Variable

A local variable is a variable created inside a function.
It only works inside that function.

‚úî You cannot use it outside the function.

‚úÖ Global Variable

A global variable is created outside any function.
It can be used anywhere in the program (inside or outside functions).

```python
# local Variable
def msg():
    choice="Hello World"
    print(choice)
msg()
print(choice) # we can not use it outside the function
```

```python
# Global Variable
def msg():
    print("inside the function"+choice)
choice="Welcome"
msg()
```

‚≠ê What is a Decorator?

A decorator is a special function that adds extra features to another function without changing its code.

Think of a decorator like:

üëâ Adding toppings on a pizza without changing the base.
üëâ Wrapping a gift without changing what‚Äôs inside.

‚≠ê Why do we use decorators?

To add extra functionality

Without modifying the original function

For cleaner and reusable code

```python
def my_decorator(func):
    def wrapper():
        print("something is happening before the function called")
        func()
        print("something is happening after the function called")
    return wrapper
@my_decorator
def say_hello():
    print("hello")

say_hello()
```

‚≠ê What is a Generator?

A generator is a special type of iterator that generates values on the fly.

It does not store all values in memory

It yields one value at a time

Useful for large data or infinite sequences

‚úÖ Key Points

Uses yield instead of return

Can be iterated using for loop or next()

Saves memory because values are generated one by one

```python
# yield keyword
def count_down(num):
    while num>0:
        yield num
        num-=1

# using the generator
for number in count_down(5):
    print(number)
```

‚≠ê What is a Lambda Function?

A lambda function is a small anonymous function (a function without a name) that can have any number of arguments, but only one expression.

It is also called anonymous function.

Syntax is shorter than normal def function.

lambda arguments: expression
arguments ‚Üí input parameters

expression ‚Üí single line calculation or operation, returned automatically



```python
add_ten=lambda x:x+10
print(add_ten(5))
```
