```python
print('''
+ADD
-SUB
*MUL
/DIV''')
num1=eval(input("Enter first number:"))
num2=eval(input("Enter second number:"))
opp=input("Enter operator: ")
if opp=="+":
    print(num1+num2)
elif opp=="-":
    print(num1-num2)
elif opp=="*":
    print(num1*num2)
elif opp=="/":
    print(num1/num2)
else:
    print("Invalid operator")
```
