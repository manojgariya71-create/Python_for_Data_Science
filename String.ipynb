```python
# if using length in string
password="@123342565"
if(len(password)<8):
    print("Less than 8 characters")
else:
    print("Greater than 8 characters")
```

```python
# iteration
s="Hello"
for char in s:
    print(char)
```

```python
s="Manoj Singh"
t=len(s)
print(t)
for a in range(t):
    print(s[a])
```

```python
# reverse
s="Manoj Singh"
t=len(s)
for a in range(t-1,-1,-1):
    print(s[a])
```

```python
s="Manoj Singh"
m=s[-1:-12:-1]
for b in m:
    print(b)
```
