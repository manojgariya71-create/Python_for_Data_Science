A tuple is an ordered, immutable (unchangeable) collection of items in Python, written inside round brackets (), and it can store multiple data types.

| Rule No. | Rule                          |
| -------- | ----------------------------- |
| 1        | Use `()` to define            |
| 2        | Ordered                       |
| 3        | Immutable                     |
| 4        | Allows duplicates             |
| 5        | Can store different types     |
| 6        | Single item tuple needs comma |
| 7        | Can be nested                 |
| 8        | Supports indexing & slicing   |


```python
a=(1,2,3,5)
print(type(a))
print(a)
b=()
print(type(b))
```

```python
# accessing tuple
a=(1,2,3,4)
print(a[0])
```

```python
# Slicing in tuples
a=(1,2,3,4,5)
print(a[0:3:1])
```
