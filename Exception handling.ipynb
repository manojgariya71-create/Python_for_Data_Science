```python
# Printing the table
a = int(input("Enter a number: "))
print(f"Multiplication of {a} is:")
for i in range(1, 11):
    print(f"{a} X {i} = {a * i}")
```

```python
try:
    num=int(input("enter a number:"))
    result=10/num
    print(f'result:{result}')
except ZeroDivisionError:
    print("you cannot divide with 0")
except ValueError:
    print("can't divide with string")
```

```python
# What is the purpose of finally block?
try:
    num=int(input("enter a number:"))
    result=10/num
    print(f'result:{result}')
except ZeroDivisionError:
    print("you cannot divide with 0")
except ValueError:
    print("can't divide with string")
finally:
    print("Always runs ")
```

```python
# What is the use of try and except?
try:
    x = int(input("Enter: "))
    print(x)
except ValueError:
    print("Invalid number")
```

```python
# What is the purpose of finally block?
try:
    a = 10
    print(b)
except:
    print("Error")
finally:
    print("Always executed")
```

```python
# What is else in exception handling?
try:
    print("Hello")
except :
    print("Error")
else:
    print("No error")
```

```python
# Multiple except blocks
try:
    a = int("abc")5
except ValueError:
    print("Value Error")
except TypeError:
    print("Type Error")
```

```python
# Catching all exceptions
try:
    a=10
    print(b)
except Exception as e:
    print("Error:", e)
```

```python
# IndexError
try:
    l=[1,2,3,4]
    print(l[5])
except Exception as a:
    print("Error",a)
finally:
    print("Always Executed")
```

```python
# Try with no except (using only finally)
try:
    print("Testing")
finally:
    print("Finally runs always")
```

```python
# Nested try-except
try:
    a = 10
    try:
        print(a / 0)
    except ZeroDivisionError:
        print("Inner division error")
except:
    print("Outer error")
```

```python
# Using pass in exception
try:
    print(10 / 0)
except ZeroDivisionError:
    pass   # ignore the error
```

```python
# Custom Exception
def check_password(password):
    if len(password) <8:
        raise Exception("Error: Password must be >=8 characters")
    print("Password is Strong")

try:
    password = input("Enter a password:")
    check_password(password)
except Exception as e:
    print("Error:",e)
```

```python
# File Handling with Try–Except–Finally
try:
    f = open("data.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File missing")
except NameError:
    print("Name Error")
finally:
    f.close()
```
