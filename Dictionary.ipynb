A dictionary in Python is a collection of key-value pairs.

Keys → unique identifiers

Values → data associated with keys

Dictionaries are unordered (before Python 3.7) and mutable.

| Feature | Rule                                      |
| ------- | ----------------------------------------- |
| Key     | Must be unique and immutable              |
| Value   | Can be anything                           |
| Mutable | Yes, can add/update/remove                |
| Order   | Python ≥3.7 → ordered, before → unordered |
| Syntax  | `{key: value, key2: value2}`              |


```python

my_dict={
    "name":"manoj",
    "age": 22,
    "marks": [22,20,24,27]
}
print(my_dict)
```

```python
# add fuction
my_dict={"fruits":["banana,orange,apple"],"category":["fruits"]}
print(my_dict)
my_dict["price"]=100
print(my_dict)
```

```python
# update
my_dict={"name":"python",
         "version":3.07}
print(my_dict)
my_dict["version"]=4.00
print(my_dict)
```

```python
# delete
my_dict={"name":"python",
         "version":3.07}
print(my_dict)
del my_dict["version"]
print(my_dict)
```

```python
# get method
profile={'name':'manoj',
         'age':22,
         'salary':100000}
age=profile.get('age','not found')
print(age)
```

```python
# keys
profile={'name':'manoj',
         'age':22,
         'salary':100000}
keys=profile.keys()
print(list(keys))
```

```python
# value
profile={'name':'manoj',
         'age':22,
         'salary':100000}
value=profile.values()
print(list(value))
all_items=profile.items()
print(list(all_items))
```

```python
# pop item  method
profile={'name':'manoj',
         'age':22,
         'salary':100000}
popped=profile.popitem()
print(popped)
print(profile)
```

```python
# pop
profile={'name':'manoj',
         'age':22,
         'salary':100000}
popped=profile.pop("name")
print(popped)
print(profile)
```

```python
# clear
profile={'name':'manoj',
         'age':22,
         'salary':100000}
print(profile)
clear=profile.clear()
print(profile)
```

```python
# nested dictionary
language={
    "python":{"name":"python","use_case":['Ai,ml,webdev,ds']}
    ,"java":{"name":"java","use_case":['App dev']}
}
print(language)
keys=language.keys()
print(list(keys))
```

```python
#loops in dict
profile={'name':'manoj',
         'age':22,
         'salary':100000}
for k in profile:
    print(k)
for k in profile.values():
    print(k)
for k in profile.items():
    print(k)
```
