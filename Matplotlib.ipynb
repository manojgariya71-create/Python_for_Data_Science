```python
import matplotlib.pylab as plt
x=['Mon','Tue','Wed','Thu','Fri']
y=[10,12,7,20,15]
plt.plot(x,y)
plt.title('Bakery Sales This Week')
plt.xlabel('Day of The Week')
plt.ylabel('Sales Per Day')
plt.show()
```

```python
# plotting - plt.plot(x,y,color='color name', linestyle='Line_style', linewidth=value, marker='marker symbol', label='label_name')
# labeling - plt.xlabel('text') or plt.ylabel('text')
# legend - Display Legend plt.legend(loc='upper left' or 'lower right')
# Axis limits - plt.xlim() or plt.ylim()
# Axis Ticks - plt.xticks() or plt.yticks()
import matplotlib.pyplot as plt
months=[1,2,3,4,5]
sales=[1000,1500,1200,1400,1800]
plt.plot(months,sales,color='red',linestyle='--',linewidth=2,marker='o',label='2025 Sale Data')
plt.xlabel('Month')
plt.ylabel('Sales Per Month')
plt.title('Monthly Sale Data Report')
plt.legend(loc='lower right')
# plt.xlim(0,6)
# plt.ylim(500,3000)
plt.xticks([1,2,3,4,5],['M1','M2','M3','M4','M5'])
plt.yticks()
plt.grid(color='blue', linestyle=':', linewidth=1)
plt.show()
```

```python
# Bar Chart
import matplotlib.pyplot as plt

product=['A','B','C','D']
sales=[1000,1500,1200,800]
plt.bar(product,sales,color='orange',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales comparison')
plt.legend()
plt.grid(color='green', linestyle=':', linewidth=1)
plt.show()
```

```python
# Pie Chart
# plt.pie(values,labels=label_list,color=color_list,)
import matplotlib.pyplot as plt
regions=['North','South','East','West']
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contributions by Region')
plt.show()
```

```python
# Histogram Chart
import matplotlib.pyplot as plt
score=[45,67,43,67,78,90,83,34,85,76,34,82,74,69,92,63,75,34,54,65,62]
plt.hist(score, bins=10,color='purple',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Histogram of Score Range')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load iris dataset
df = sns.load_dataset('iris')# Visualize pairwise relationships
sns.pairplot(df, hue='species', palette='husl')
plt.show()# Display basic statistics
print(df.groupby('species').mean())
```
