```python
#While Loop
i=1
while(i<5):
    print("Hello World")
    i+=1
```

```python
#for loop
l=[1,2,3,4,5]
for i in l:
    print(i + 10)
```

```python
# infinite loop in while
while(True):
    print("Hello World")
```

```python
#range

for i in range(1,11,1):
    print(i)
```

```python
#convert range into list
a=list(range(1,11,2))
print(a)
```

```python
# coverting into tuple

a=tuple(range(1,11,2))
print(a)
```

```python
# nested loop

for i in range(1,3):
    for j in range(3,6):
        print(i,j)
```

```python
# Reverse loop
for i in range(10,0,-1):
    print(i)
```

```python
l=list(range(20,0,-1))
print(l)
```

```python
for i in range(5):
    print(i)
```

```python
fruits=["apple","mango","banana"]
for fruits in fruits:
    print(fruits)
```

```python
# loop through string
for char in "Python":
    print(char)
```

```python
# using break statement in loop
for i in range(1,10,1):
    if i==5:
        break
    print(i)
```

```python
# using continue statement in loop
for i in range(1,10,1):
    if i==5:
        continue
    print(i)
```

```python
# Even and Odd
for i in range(1,11,1):
    if i%2==0:
        print("Even",i)
    else:
        print("Odd",i)
```

```python
# skip the num using input
start=int(input("Enter the start = "))
end=int(input("Enter the stop = "))

skip=int(input("enter the number"))

for i in range(start,end):
    if i==skip:
        continue
    print(i)
```
